```python
import numpy as np
import matplotlib.pyplot as pt
import scipy as ci
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

<font size=5> Problem 1 <font>

```python
sns.set(font_scale=1.5)
rs=10;
C=np.arange(22,221,2)*np.pi
r=C/(2*np.pi) ## Coordinate Radius
ds=[]
dh=[]
H=[]
for i in range(0,len(C)):
    ds.append(1/np.sqrt(1-rs/r[i]))  ##  Length for each Segment 's'
    dh.append(np.sqrt(ds[i]**2-1))  ## Change in height of each loop
    H=np.append(H,sum(dh))  ## Absolute Height of each loop

t=np.arange(0,2*np.pi+0.1,0.1)
x=np.cos(t)
y=np.sin(t)  ## Paramatrizing 

fig=pt.figure(figsize=(10,5))
ax=fig.gca(projection='3d')

for i in range(0,len(C)):
    ax.plot(r[i]*x,r[i]*y,H[i],'-')
pt.title('Loops')
ax.set_xlim(-100,100)
ax.set_ylim(-100,100)
ax.set_zlim(0,80)
pt.show()
```

<font size=5> Problem 2 <font>

```python
v=0.1
ds=[]
S=[]
dt=[]
T=[]

for i in range(0,len(C)):
    ds.append(1/np.sqrt(1-rs/r[i]))  ## Rope Length for each Segment
    S=np.append(S,sum(ds))  ## Total lenght of rope    
    dt.append(ds[i]/v*np.sqrt(1-rs/r[i]))  ## Time increment near mass- 
                                            # -for every segment traveled
    T.append(sum(dt)) ## Total time spent near mass

Cable=S[-2] ## Length of cable in total, excluding the last segment on the 100th loop

print('stationary clock round trip time: '+str(Cable/v*2-ds[-1]/v*2)) 
## Distance/Velocity = Total time in rest frame
### Above expression excludes time at 100th loop 
print('descending clock round trip time: '+str(T[-1]*2))
print('coordinate radius difference of initial and final positions: '+str(len(C)-1))
print('cable length required is: '+str(Cable))
```
