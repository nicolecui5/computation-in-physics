<font size=5>Problem 1: Evaluating $\pi$ </font>

```python
sns.set(font_scale=2)
sns.set_style("white")
def factorial(n):        ### Making algorithm for Factorial
    tmp=[]
    for i in range(1,n+1):
        tmp.append(i) 
    tmp2=1
    for i in tmp:
        tmp2=tmp2*i
    return tmp2

Pi=[]
n=3                      ## Number of Iterations
for i in range(1,n+1):
    tmp=[]
    for k in range(0,i):
        tmp.append(factorial(4*k)*(1103+26390*k)/((factorial(k))**4*396**(4*k)))
    pi=sum(tmp)
    Pi.append(1/(2*np.sqrt(2)/9801*pi))
    
pt.figure(figsize=(20,5))   ## Plotting calculated value of Pi
pt.subplot(22 1)
r=np.ones(n)
r=r*np.pi
x=np.linspace(1,n,n)
pt.plot(x,Pi,'o',label=r'$\pi$',markersize=13,color='black')
pt.axhline(np.pi,color='black',label=r'Real $\pi$')
pt.xlabel('Terms')
pt.xlim(0,n+1)
pt.ylim(np.pi-0.0000001,np.pi+0.0000001)
pt.title(r'Value of $\pi$ Keeping Different Numbers of Terms',y=1.1)
pt.legend()

pt.subplot(122) ## Plotting differences from calculated and real value of Pi
pt.plot(x,r-Pi,'o',label='Difference', markersize=13,color='black')
pt.axhline(0,label='Difference of 0',color='black')
pt.xlabel('Terms')
pt.xlim(0,n+1)
pt.ylim(-0.0000001,0.0000001)
pt.title(r'Difference Between Calculated and Real $\pi$',y=1.1)
pt.legend()
pt.show()
```

<font size=4> In the first plot, the dots indicates the value of $\pi$ calculated after keeping 1,2,and 3 terms and the horizontal line is the true value of $\pi$, and it's clear that the value of $\pi$ calculated using more terms is more accurate. The second plot is the diffrence between calculated number and the real value of $\pi$ when keeping different terms, and it's obvious that the difference is converging to 0 quickly. <font>

<font size=5>Problem 2: Integration<font>

```python
c=299792458
t=10                       ## time increment = 10 sec
a=9.81
vship=c*0.5                ## Starting at half speed of light!
tmp_x=[]
temp_t=[]
X=[]
V=[]
T=[] 
T_ship=86400               ## 1 days duration
for i in range(0,T_ship//10):
    v=(a*t*i+vship)/(1+a*t*i*vship/c**2)   ## updating velocity
    dt=10/np.sqrt(1-(v/c)**2)
    dx=v*dt
    tmp_x.append(dx)
    temp_t.append(dt)
    V.append(v)             
    X.append(sum(tmp_x))    ## integrating distance traveled
    T.append(sum(temp_t))   ## integrating time observed on earth
```

```python
print("ship time (sec)="+str(86400))
print("earth time (sec)="+str(np.round(T[-1],2)))
print("ship speed (c)="+str(V[-1]/(c)))
print("ship distance (Ly)="+str((X[-1]/(c*365*86400))))
```
