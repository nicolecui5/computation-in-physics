```python
import numpy as np
import matplotlib.pyplot as pt
import scipy as ci
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

<font size=5> Problem 1<font>

```python
sns.set(font_scale=1.3)
sns.set_style('white')
r=[]
s=[]
h=[]
H=[]
for i in range(0,49):
    r.append((i+1)*0.02) ## Radius of loop
    s.append(0.02/np.sqrt(1-r[i]**2)) ## Length of String
    h.append(np.sqrt(s[i]**2-0.02**2)) ## Height Increment
    H.append(sum(h)) ## Sum to get the height of ith loop
    
fig=pt.figure(figsize=(10,7))
ax=fig.gca(projection='3d')

t=np.arange(0,2*np.pi,0.1)
x=np.cos(t)
y=np.sin(t)
for i in range(0,49):
    pt.plot(r[i]*x,r[i]*y,H[i])
    
ax.set_xlabel('Radius',labelpad=10)
ax.set_zlabel('Height',labelpad=5)
pt.title('Rings with Increasing Heights')
pt.show()
```

```python
sns.set(font_scale=1.3)
sns.set_style('white')
r=[]
s=[]
h=[]
H=[]
for i in range(0,49):
    r.append((i+1)*0.02) ## Radius of loop
    s.append(0.02/np.sqrt(1-r[i]**2)) ## Length of String
    h.append(np.sqrt(s[i]**2-0.02**2)) ## Height Increment
    H.append(sum(h)) ## Sum to get the height of ith loop
    
fig=pt.figure(figsize=(10,10))
ax=fig.gca(projection='3d')

t=np.arange(0,2*np.pi,0.1)
x=np.cos(t)
y=np.sin(t)
for i in range(0,49):
    pt.plot(r[i]*x,r[i]*y,H[i],color='royalblue')

########## Drawing Sphere
r=1   ## raius=1
theta=np.arange(0,np.pi,0.01)
phi=np.arange(0,2*np.pi,0.01)
theta,phi=np.meshgrid(theta,phi)

xc=r*np.sin(theta)*np.cos(phi) 
yc=r*np.sin(theta)*np.sin(phi)
zc=r*np.cos(theta)

ax.plot_surface(xc,yc,zc+1,alpha=0.4,cmap='rainbow')
ax.set_xlabel('Radius',labelpad=10)
ax.set_zlabel('Height',labelpad=10)
pt.title('Sphere with R=1 and Rings')
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(0,2)
pt.show()
```
