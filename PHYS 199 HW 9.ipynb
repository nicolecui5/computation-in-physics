```python
## Just making selections using final coordinates. Exhausted........
    
######
#         self.k0_trajectory_start      = np.array([0., 0., 0.])
#         self.k0_trajectory_stop       = np.array([0., 0., 0.]) = z_decay
#         self.piplus_trajectory_stop   = np.array([0., 0., 0.])
#         self.piminus_trajectory_stop  = np.array([0., 0., 0.])
#####  Calo:1.5*1.5m   hole: 0.5*0.5m

import week09_particledetectorObject as ParticleDetectorClassDefinition
import week09_histogramObject as HistogramCLassDefinition
detector = ParticleDetectorClassDefinition.ParticleDetector()     

number_of_kaons_total = 100000

both_cal=0
both_hole=0
hole_cal=0
hole_nocal=0
cal_nocal=0
both_nocal=0
kaon_far=0

Out=1.5/2
In=0.5/2

for i in range(0, number_of_kaons_total):
    detector.generate_one_kaon_decay()

    ## both at cal:
    if      abs(detector.piplus_trajectory_stop[0]) < Out \
        and abs(detector.piplus_trajectory_stop[1]) < Out\
        and(abs(detector.piplus_trajectory_stop[0]) > In \
        or  abs(detector.piplus_trajectory_stop[1]) > In)\
        and abs(detector.piminus_trajectory_stop[0]) < Out \
        and abs(detector.piminus_trajectory_stop[1]) < Out\
        and(abs(detector.piminus_trajectory_stop[0]) > In \
        or abs(detector.piminus_trajectory_stop[1]) > In):\
        both_cal+=1
        
    ## both at hole:
    if      abs(detector.piplus_trajectory_stop[0]) < In\
        and abs(detector.piplus_trajectory_stop[1]) < In\
        and abs(detector.piminus_trajectory_stop[0]) < In\
        and abs(detector.piminus_trajectory_stop[1]) < In:
        both_hole+=1
                
    ## one hole one cal: 
    if     (abs(detector.piplus_trajectory_stop[0]) < In\
        and abs(detector.piplus_trajectory_stop[1]) < In\
        and abs(detector.piminus_trajectory_stop[0]) < Out \
        and abs(detector.piminus_trajectory_stop[1]) < Out\
        and(abs(detector.piminus_trajectory_stop[0]) > In \
        or abs(detector.piminus_trajectory_stop[1]) > In))\
        or (abs(detector.piminus_trajectory_stop[0]) < In\
        and abs(detector.piminus_trajectory_stop[1]) < In\
        and abs(detector.piplus_trajectory_stop[0]) < Out \
        and abs(detector.piplus_trajectory_stop[1]) < Out\
        and(abs(detector.piplus_trajectory_stop[0]) > In \
        or abs(detector.piplus_trajectory_stop[1]) > In)):
        hole_cal+=1
    
    ## one hole one out of cal: 
    if     (abs(detector.piplus_trajectory_stop[0]) < In\
        and abs(detector.piplus_trajectory_stop[1]) < In\
        and abs(detector.piminus_trajectory_stop[0]) > Out\
        and abs(detector.piminus_trajectory_stop[1]) > Out) \
        or (abs(detector.piminus_trajectory_stop[0]) < In\
        and abs(detector.piminus_trajectory_stop[1]) < In\
        and abs(detector.piplus_trajectory_stop[0]) > Out\
        and abs(detector.piplus_trajectory_stop[1]) > Out):
        hole_nocal+=1
        
    ## one at cal one out of cal:
    if     (abs(detector.piminus_trajectory_stop[0]) < Out \
        and abs(detector.piminus_trajectory_stop[1]) < Out\
        and(abs(detector.piminus_trajectory_stop[0]) > In \
        or abs(detector.piminus_trajectory_stop[1]) > In)\
        and abs(detector.piplus_trajectory_stop[0]) > Out\
        and abs(detector.piplus_trajectory_stop[1]) > Out)\
        or (abs(detector.piplus_trajectory_stop[0]) < Out \
        and abs(detector.piplus_trajectory_stop[1]) < Out\
        and(abs(detector.piplus_trajectory_stop[0]) > In \
        or abs(detector.piplus_trajectory_stop[1]) > In)\
        and abs(detector.piminus_trajectory_stop[0]) > Out\
        and abs(detector.piminus_trajectory_stop[1]) > Out):
        cal_nocal+=1

    ## both out of cal:
    if      abs(detector.piminus_trajectory_stop[0]) > Out\
        and abs(detector.piminus_trajectory_stop[1]) > Out \
        and abs(detector.piplus_trajectory_stop[0]) > Out\
        and abs(detector.piplus_trajectory_stop[1]) > Out:
        both_nocal+=1
    
    ## kaon too far:
    if detector.z_decay>45:
        kaon_far+=1
        
print(' ')
print('================ Fractions: ==================')
print(' ')
print('both at calorimeter:  '+str(both_cal/number_of_kaons_total*100)+'%')
print('both in hole:  '+str(both_hole/number_of_kaons_total*100)+'%')
print('one in hole, one at calorimeter:  '+str(hole_cal/number_of_kaons_total*100)+'%')
print('one in hole, one out of calorimeter:  '+str(\
       hole_nocal/number_of_kaons_total*100)+'%')
print('one at calorimter, one out of calorimeter:  '+\
      str(cal_nocal/number_of_kaons_total*100)+'%')
print('both out of calorimeter:  '+str(both_nocal/\
   number_of_kaons_total*100)+'%')
print('kaon gone too far:  '+str(kaon_far/number_of_kaons_total*100)+'%')
print('fraction of at least one in hole: '+str((both_hole+\
        hole_cal+hole_nocal)/number_of_kaons_total*100)+'%')
```
