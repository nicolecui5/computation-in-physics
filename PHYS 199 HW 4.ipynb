```python
import numpy as np
import matplotlib.pyplot as pt
import scipy as ci
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

<font size=5>Problem 1: Displacement and Velocity<font>

```python
# dx/dt=v , dv/dt=-A*v-B*sin(x)+C*sin(W*t)
sns.set(font_scale=1.5)
def Oscillate(a,b,c,w,time):
    dt=0.01 ## Increment

    def F(x,v,t): ## 1st order Equation
        A=a
        B=b
        C=c
        W=w
        return -A*v-B*np.sin(x)+C*np.sin(W*t)
    ## Innitial Conditions
    t0=0
    x0=0
    v0=0.1
    ## Lists which are used to plot
    T=[]
    X=[]
    V=[]
    for i in range(0,round(time/dt)):
        dx0=v0*dt    ## Algorithm on Lecture Notes
        dv0=F(x0,v0,t0)*dt

        x0_1=x0+dx0/2
        v0_1=v0+dv0/2

        dx_i=v0_1*dt
        dv_i=F(x0_1,v0_1,t0+dt)*dt

        x1=x0+dx_i   
        v1=v0+dv_i

        x0=x1  ## Updating X,V,T
        v0=v1
        t0=t0+dt

        T.append(t0)
        X.append(x0)
        V.append(v0)
    pt.figure(figsize=(15,5))
    pt.subplot(121)
    pt.plot(T,X)
    pt.xlabel('t')
    pt.ylabel('posiiton')
    pt.title('A='+str(a)+', B='+str(b)+', C='+str(c)+', W='+str(w))
    pt.subplot(122)
    pt.plot(T,V)
    pt.xlabel('t')
    pt.ylabel('velocity')
    pt.title('A='+str(a)+', B='+str(b)+', C='+str(c)+', W='+str(w))
```

```python
## Version 1
Oscillate(0,0.1,0,0,120)
```

```python
## Version 2
Oscillate(0.1,1,0.1,2,120)
```

```python
## Version 3
Oscillate(0.1,1,2,1.2,500)
```
